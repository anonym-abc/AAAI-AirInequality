```python
import pandas as pd
```

```python
city_df = pd.read_csv("AirQuality/Dataset/Ground_Truth_2023_Final.csv")
ndvi_df = pd.read_csv("AirQuality/RQ3/Dataset/city_monthly_ndvi_2023.csv")
population_df = pd.read_csv("AirQuality/RQ3/Dataset/Population.csv")

print(city_df.shape)
print(ndvi_df.shape)
print(population_df.shape)

print(city_df.columns)
print(ndvi_df.columns)
print(population_df.columns)
```

```python
population_df.head()
```

```python
for df in [city_df, ndvi_df, population_df]:
    df['city'] = df['city'].str.lower().str.strip()
    df['state'] = df['state'].str.lower().str.strip()

city_df['YearMonth'] = pd.to_datetime(city_df['YearMonth'])
city_df['month'] = city_df['YearMonth'].dt.strftime('%Y-%m') 

merged_df = city_df.merge(ndvi_df, on=['city', 'state', 'month'], how='left')

merged_df = merged_df.merge(population_df[['city', 'state', 'population']], on=['city', 'state'], how='left')

final_df = merged_df[[
    'city', 'state', 'month', 'AT', 'latitude', 'longitude', 'PM2.5', 'avg_ndvi', 'max_ndvi', 'min_ndvi', 'population'
]]

final_df.head()
```

```python
final_df.to_csv("AirQuality/RQ4/Dataset/GroundTruth_With_Aux_Data.csv",index=False)
```
